```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('train-HW_1-sid_13.csv')
dataP = pd.read_csv('test-HW_1-sid_13.csv')
#data.info()
#dataP.info()
print(data.describe())
```

```python
plt.figure(figsize=(10,5))
c= data.corr()
sns.heatmap(c)
```

```python
data.corr()['target']
```

```python
from sklearn import preprocessing 
norm_scel=preprocessing.MinMaxScaler()
std_scel=preprocessing.StandardScaler()
columns = list(data.columns)

print(columns)
```

```python
for column in columns:
    value = data[column].mean()
    data[column].fillna(value,inplace=True)

data.info()
```

```python
df_norm = pd.DataFrame(norm_scel.fit_transform(data))
df_norm.head(20)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
df_norm[range(0,16)].head(20)
spisok = list(df_norm.columns)
print(spisok)
#df_norm[16].head(20)
#df_norm.head(20)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_norm[spisok[:-1]], data['target'], test_size=0.2)
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)
MSE = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

coef=regr.coef_
print('MSE=',MSE,'R_2=',r2)
print("Coefficients: ", coef)
x = range(1,len(coef)+1)  
plt.plot(x,coef) #[-50:50] не важные, остальные важные
plt.plot(x,[50]*len(x),linestyle='--',color ='k')
plt.plot(x,[-50]*len(x),linestyle='--',color ='k')
plt.xlabel('№ Feature')
plt.ylabel('ves')
imp = []
for i in range(0,len(coef)):
    if abs(coef[i])>50:
        imp.append(i)
print('Important coefs have № : ',imp)
```

```python
from sklearn.linear_model import Ridge
MSE_l = []
R2_l = []
for n in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(df_norm[spisok[:-1]], data['target'], test_size=0.2)
    regR = linear_model.Ridge(alpha=0.5)
    regR.fit(X_train,y_train)
    y_predR = regR.predict(X_test)
    MSE_l.append(mean_squared_error(y_test, y_predR))
    R2_l.append(r2_score(y_test, y_predR))
print('MSE average ', np.mean(MSE_l))
print('R2_avg ',np.mean(R2_l))
```

```python
from sklearn.linear_model import Lasso
MSE_l = []
R2_l = []
for n in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(df_norm[spisok[:-1]], data['target'], test_size=0.2)
    regL = linear_model.Lasso(alpha=0.1)
    regL.fit(X_train,y_train)
    y_predL = regL.predict(X_test)
    MSE_l.append(mean_squared_error(y_test, y_predL))
    R2_l.append(r2_score(y_test, y_predL))
print('MSE average ', np.mean(MSE_l))
print('R2_avg ',np.mean(R2_l))
```
